```python
import pandas as pd
#import pandas_datareader as dr
import numpy as np
import warnings
import matplotlib.pyplot as plt
```

```python
### Reading in AAPL stock data and cleaning ###
aapl = pd.read_csv("data/AAPL_1min_sample cleaned.txt", names = ["Time", "Open", "High", "Low", "Close", "Volume"])
aapl.insert(0, 'Day', pd.to_datetime(aapl['Time']).dt.date)
aapl.insert(1, 'Hour', pd.to_datetime(aapl['Time']).dt.time)
aapl = aapl.drop(['Time'], axis=1).set_index('Day')
aapl
```

```python
aapl_days = []
for i in aapl.index.unique():
    aapl_days.append(aapl.loc[i])
```
