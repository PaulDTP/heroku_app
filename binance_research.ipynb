```python
from mysite.logger import log_status
import os
from dotenv import load_dotenv
from binance.spot import Spot as Client
from binance.websocket.spot.websocket_stream import SpotWebsocketStreamClient
```

```python
load_dotenv()
api_key = os.getenv("TEST_API_KEY2")
api_secret = os.getenv("TEST_API_SECRET2")
```

```python
client = Client(api_key, api_secret, base_url='https://testnet.binance.vision')
response = client.new_listen_key()
log_status('info', f"Received listen key: {response}")
```

```python
ws_client = SpotWebsocketStreamClient(
    stream_url="wss://testnet.binance.vision", on_message=message_handler
)
ws_client.user_data(listen_key=response["listenKey"])
```

```python
ws_client.user_data(
    response["listenKey"], action=SpotWebsocketStreamClient.ACTION_UNSUBSCRIBE
)
```

```python
ws_client.stop()
```

```python
from binance.websocket.spot.websocket_api import SpotWebsocketAPIClient

def message_handler(_, message):
    log_status('info', message)
    
def on_close(_, close_event):
    print("closing")
    print(close_event)

my_client = SpotWebsocketAPIClient(
    stream_url="wss://testnet.binance.vision/ws-api/v3",
    api_key=api_key,
    api_secret=api_secret,
    on_message=message_handler,
    on_close=on_close,
)
my_client.user_data_start()
```

```python
my_client.stop()
```
